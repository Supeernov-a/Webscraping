# Web-Scraping

The goal of this notebook is to extract information data from "box office mojo"(the below picture), using Web Scraping

image.png

## Libraries for WebScraping

```python
import requests
from bs4 import BeautifulSoup
```

```python
Web_url ="https://www.boxofficemojo.com/chart/top_lifetime_gross/?area=XWW"
r = requests.get(Web_url)
# Parse HTML Code
soup = BeautifulSoup(r.content, 'html.parser')
soup.prettify()
```

**Obtaining the name of movie**

```python
movie = soup.find_all("td", {"class":"a-text-left mojo-field-type-title"})
Movies = []
for i in movie:
  Movies.append(i.get_text())
print(Movies)
```

**Obtaining the year of release**

```python
year = soup.find_all("td", {"class":"a-text-left mojo-field-type-year"})
Years = []
for i in year:
  Years.append(i.get_text())
print(Years)
```

**Obtaining the NetWorth of the movie**

```python
networth = soup.find_all("td", {"class":"a-text-right mojo-field-type-money"})
NetWorth = []
for i in networth:
  NetWorth.append(i.get_text().replace("$","").replace(",",""))
print(NetWorth)
```

**Creating a dataframe with the above data**

## Data Frame

```python
import pandas as pd

data = pd.DataFrame()
data["Movie"]= Movies
data["Year"]= Years
data["NetWorth"]= NetWorth
data
```

**Info about the data**

```python
data.info()
```

```python
data["NetWorth"] = pd.to_numeric(data["NetWorth"]) 
data.info()
```

```python
data.to_csv("Mydata.csv", index=False)
data.to_excel("excel.xlsx")
```

## Data Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

**Amount of Money per Year**

```python
col= ['2009', '2019', '2022', '1997', '2015', '2018', '2021', '2012',
       '2011', '2017', '2013', '2016', '2003', '1993', '2014', '2010',
       '2006', '1999', '2001', '2008', '1994', '2007', '2002', '2004',
       '2005', '1996', '1982', '1977', '1998', '2000']
col= sorted(col)
NewperYear = []
for i in col:
  NewperYear.append(data.loc[data.Year==i,"NetWorth"].sum())


plt.figure(figsize= (10,15))
sns.barplot(x=col,y=NewperYear)
plt.xlabel("Year")
plt.ylabel("NetWorth Total")
plt.xticks(rotation =90)
plt.title("Amount of Money per Year")

```
